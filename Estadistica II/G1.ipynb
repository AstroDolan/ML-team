```python
import math
import scipy.stats as st
```

###  Ejercicio 1 (Control de Calidad en Café)
Aquí utilizamos st.norm.cdf(z), que es el equivalente exacto a buscar el valor $Z$ en la tabla de probabilidad acumulada (el área hacia la izquierda de la curva).

```python
# Datos del problema
mu_1 = 500
sigma_1 = 10
n_1 = 25
x_bar_1 = 496

# Cálculos
error_estandar_1 = sigma_1 / math.sqrt(n_1)
z_1 = (x_bar_1 - mu_1) / error_estandar_1

# cdf (Cumulative Distribution Function) calcula P(Z < z)
prob_1 = st.norm.cdf(z_1)

print("--- Ejercicio 1 ---")
print(f"Valor Z: {z_1}")
print(f"Probabilidad P(X < 496): {prob_1:.4f} ({prob_1*100:.2f}%)")
```

###  Ejercicio 3 (Memoria y Aprendizaje)
Como nos piden la probabilidad de que el promedio sea superior a 3 palabras, necesitamos calcular la cola derecha de la distribución. En Python, en lugar de hacer 1 - cdf, podemos usar directamente sf (Survival Function), que calcula el área a la derecha de $Z$.

```python
# Datos del problema
mu_3 = 2.75
sigma_3 = 1.20
n_3 = 64
x_bar_3 = 3

# Cálculos
error_estandar_3 = sigma_3 / math.sqrt(n_3)
z_3 = (x_bar_3 - mu_3) / error_estandar_3

# sf (Survival Function) calcula P(Z > z)
prob_3 = st.norm.sf(z_3)

print("--- Ejercicio 3 ---")
print(f"Valor Z: {z_3:.4f}")
print(f"Probabilidad P(X > 3): {prob_3:.4f} ({prob_3*100:.2f}%)")
```

### Ejercicio 4 (Duración de Baterías)
Para encontrar un límite de intervalo de confianza, necesitamos el proceso inverso: a partir de una probabilidad (95%), encontrar el valor $Z$ crítico. Para eso usamos ppf (Percent Point Function).

```python
# Datos del problema
sigma_4 = 1.5
n_4 = 9
x_bar_4 = 6.2
nivel_confianza = 0.95

# Para un 95% central, las colas suman 5% (2.5% cada una).
# Buscamos el Z para el percentil 0.975 (0.95 + 0.025)
z_critico_4 = st.norm.ppf(1 - (1 - nivel_confianza) / 2)

error_estandar_4 = sigma_4 / math.sqrt(n_4)
margen_error = z_critico_4 * error_estandar_4
limite_inferior = x_bar_4 - margen_error

print("--- Ejercicio 4 ---")
print(f"Z crítico: {z_critico_4:.3f}")
print(f"Límite Inferior: {limite_inferior:.2f} horas")
```

### Ejercicio 5 (Proporción de Fallas)
 Este ejercicio usa la distribución muestral de una proporción. La lógica de la estandarización $Z$ es la misma, solo cambia cómo calculamos el error estándar.

```python
# Datos del problema
p_5 = 0.10
n_5 = 100
p_hat_5 = 0.13

# Cálculo del error estándar para proporciones
error_estandar_5 = math.sqrt((p_5 * (1 - p_5)) / n_5)
z_5 = (p_hat_5 - p_5) / error_estandar_5

# Probabilidad de que sea mayor (cola derecha)
prob_5 = st.norm.sf(z_5)

print("--- Ejercicio 5 ---")
print(f"Valor Z: {z_5}")
print(f"Probabilidad P(p > 0.13): {prob_5:.4f} ({prob_5*100:.2f}%)")
```
