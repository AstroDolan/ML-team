### Comparación de Estimadores por Error Cuadrático Medio (ECM)
¿Cuándo usar este bloque?
Úsalo en ejercicios como el 2 (Termómetros) o el 8 (GPS). Te servirá cada vez que te den dos o más instrumentos/estimadores, te entreguen su sesgo y su varianza, y te pidan decidir cuál es el mejor. El mejor siempre es el que tiene el menor ECM.

```python
def calcular_ecm(varianza, sesgo):
    """Calcula el Error Cuadrático Medio: ECM = Var + Sesgo^2"""
    return varianza + (sesgo ** 2)

print("--- Comparación por ECM ---")
# Datos Instrumento A
var_A = 0.05
sesgo_A = 0.2
ecm_A = calcular_ecm(var_A, sesgo_A)

# Datos Instrumento B
var_B = 0.15
sesgo_B = 0.0
ecm_B = calcular_ecm(var_B, sesgo_B)

print(f"ECM Instrumento A: {ecm_A:.4f}")
print(f"ECM Instrumento B: {ecm_B:.4f}")

if ecm_A < ecm_B:
    print("Conclusión: Es preferible el Instrumento A (Menor ECM).")
elif ecm_B < ecm_A:
    print("Conclusión: Es preferible el Instrumento B (Menor ECM).")
else:
    print("Conclusión: Ambos instrumentos son igualmente buenos según su ECM.")
```

### Cálculo de Eficiencia Relativa
¿Cuándo usar este bloque?
Úsalo en problemas como el Ejercicio 3 (Análisis de Suelos). Te darán la varianza de dos metodologías distintas (por ejemplo, medir una vez vs. medir 5 veces) y te pedirán comparar qué tan eficiente es una respecto a la otra.

```python
print("--- Eficiencia Relativa ---")
# La eficiencia relativa compara Varianzas. 
# Si quieres saber qué tan eficiente es el método 2 respecto al 1, divides Var(1) / Var(2).

var_metodo_simple = 0.60
var_metodo_promedio = 0.12

eficiencia_relativa = var_metodo_simple / var_metodo_promedio

print(f"Varianza método simple: {var_metodo_simple}")
print(f"Varianza método promedio: {var_metodo_promedio}")
print(f"Eficiencia Relativa: {eficiencia_relativa:.2f}")
print(f"Conclusión: El método del promedio es {eficiencia_relativa} veces más eficiente.")
```

### Estimación por Método de los Momentos (MM)
¿Cuándo usar este bloque?
Úsalo en los ejercicios 4 (Flujo de vehículos - Poisson) y 7 (Tamaño de peces - Uniforme). Este método se basa en igualar la media teórica de una distribución con la media real que obtuviste en tu muestra de datos.
- Para Poisson: $\lambda = \bar{x}$
- Para Uniforme $U(0, \theta)$: $\frac{\theta}{2} = \bar{x} \implies \theta = 2\bar{x}$

```python
print("--- Método de los Momentos (MM) ---")

# 1. Ingresar los datos de la muestra
# Ejemplo usando los datos del Ejercicio 7 (Peces)
muestra = np.array([12, 18, 15, 20, 10])

# 2. Calcular el primer momento muestral (la media)
media_muestral = np.mean(muestra)
print(f"Media de la muestra (x_bar): {media_muestral}")

# 3. Calcular parámetros según la distribución indicada en la prueba:
distribucion = "Uniforme" # Cambia esto a "Poisson", "Uniforme" o "Exponencial" en la prueba

if distribucion == "Poisson":
    # En Poisson, E[X] = lambda
    lambda_estimado = media_muestral
    print(f"Estimación MM para Lambda: {lambda_estimado}")

elif distribucion == "Uniforme":
    # En Uniforme(0, theta), E[X] = theta / 2 -> theta = 2 * E[X]
    theta_estimado = 2 * media_muestral
    print(f"Estimación MM para Theta: {theta_estimado}")

elif distribucion == "Exponencial":
     # En Exponencial, E[X] = 1 / lambda -> lambda = 1 / E[X]
     lambda_exp_estimado = 1 / media_muestral
     print(f"Estimación MM para Lambda (Exp): {lambda_exp_estimado:.4f}")
```

### Varianza de una Combinación Lineal (Cartera de Inversiones)
¿Cuándo usar este bloque?
Úsalo en problemas como el Ejercicio 5, donde tienes un estimador que se construye sumando (o restando) variables que tienen distintas ponderaciones (pesos o porcentajes), asumiendo independencia entre ellas. La regla de oro es que los pesos o coeficientes se elevan al cuadrado al entrar a la varianza.

```python
print("--- Varianza de una Combinación Lineal ---")
# Modelo: R = a*X + b*Y
# Fórmula teórica: Var(R) = (a^2)*Var(X) + (b^2)*Var(Y)

# Coeficientes (los números que acompañan a las variables)
peso_A = 0.6
peso_B = 0.4

# Varianzas individuales
var_A_ind = 2.0
var_B_ind = 5.0

# Cálculo aplicando la propiedad
varianza_total = (peso_A**2)*var_A_ind + (peso_B**2)*var_B_ind

print(f"Varianza aportada por A: {(peso_A**2)*var_A_ind:.4f}")
print(f"Varianza aportada por B: {(peso_B**2)*var_B_ind:.4f}")
print(f"Varianza Total Estimada: {varianza_total:.4f}")
```
